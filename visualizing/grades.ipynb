```python
import pandas as pd
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
```

```python
student_dataset = pd.read_csv('./students.data', index_col=0)
student_dataset.head(5)
```

```python
my_series = student_dataset.G3
my_dataframe = student_dataset[['G3', 'G2', 'G1']]
```

```python
my_series.plot.hist(alpha=0.5)
my_dataframe.plot.hist(alpha=0.5)
```

```python
my_series.plot.hist(alpha=0.5, normed=True)
my_dataframe.plot.hist(alpha=0.5, normed=True)
```

```python
# student_dataset.plot.scatter(x='studytime', z='absences', y='failures') have to use matplotlib directly
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
plt.suptitle("Failures related to studytime and absences")
```

```python
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Study Time')
ax.set_ylabel('Failures')
ax.set_zlabel('Absences')
ax.scatter(student_dataset.studytime, student_dataset.failures, student_dataset.absences, c='r', marker='.')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Final Grade')
ax.set_ylabel('First Grade')
ax.set_zlabel('Daily Alcohol')
ax.scatter(student_dataset.G1, student_dataset.G3, student_dataset['Dalc'], c='r', marker='.')
plt.show()
```
